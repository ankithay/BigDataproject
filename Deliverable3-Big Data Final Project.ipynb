### Imputation 
##### Imputing categorical data with most frequent value
##### Imputing numerical data with mean and ffill values

```python
mode = ''.join(df1['tests_units'].mode())
df1['tests_units'].fillna(mode, inplace=True)
df1['tests_units']
```

```python
df1.fillna(df1.mean())
```

```python
df1["total_cases"] = df1["total_cases"].fillna(df1["total_cases"].mean())
df1['positive_rate']=df1["positive_rate"].fillna(df1["positive_rate"].mean())
df1['population_density']=df1["population_density"].fillna(df1["population_density"].mean())
```

```python
df2=df1[['location','date','total_cases','total_deaths','positive_rate','population_density']]
df2["date"]=pd.to_datetime(df2["date"])
```

```python
datewise=df2.groupby(["date"]).agg({"total_cases":'sum',"total_deaths":'sum'})
```

## Machine learning
* We've used Linear regression model for predicting total cases
* Train and valid datasets are created from the dataset


```python
datewise["Days Since"]=datewise.index-datewise.index[0]
datewise["Days Since"]=datewise["Days Since"].dt.days
```

```python
train_ml=datewise.iloc[:int(datewise.shape[0]*0.95)]
valid_ml=datewise.iloc[int(datewise.shape[0]*0.95):]
model_scores=[]
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.metrics import mean_squared_error,r2_score
lin_reg=LinearRegression(normalize=True)
```

```python
lin_reg.fit(np.array(train_ml["Days Since"]).reshape(-1,1),np.array(train_ml["total_cases"]).reshape(-1,1))
```

```python
prediction_valid_linreg=lin_reg.predict(np.array(valid_ml["Days Since"]).reshape(-1,1))
```

### Evaluation
* For evaluation of linear regression model we used Root mean Square Error

```python
model_scores.append(np.sqrt(mean_squared_error(valid_ml["total_cases"],prediction_valid_linreg)))
print("Root Mean Square Error for Linear Regression: ",np.sqrt(mean_squared_error(valid_ml["total_cases"],prediction_valid_linreg)))
```

```python
plt.figure(figsize=(11,6))
prediction_linreg=lin_reg.predict(np.array(datewise["Days Since"]).reshape(-1,1))
linreg_output=[]
for i in range(prediction_linreg.shape[0]):
    linreg_output.append(prediction_linreg[i][0])

fig=go.Figure()
fig.add_trace(go.Scatter(x=datewise.index, y=datewise["total_cases"],
                    mode='lines+markers',name="Train Data for total Cases"))
fig.add_trace(go.Scatter(x=datewise.index, y=linreg_output,
                    mode='lines',name="Linear Regression Best Fit Line",
                    line=dict(color='black', dash='dot')))
fig.update_layout(title="Confirmed Cases Linear Regression Prediction",
                 xaxis_title="Date",yaxis_title="total cases",legend=dict(x=0,y=1,traceorder="normal"))
fig.show()
```

   ### Optimization
   * As we can see from above root mean squared error for the linear regression is quite large
   * To decrease the root mean squared erroe and optimise the model, we've used polynomial features

```python
train_ml=datewise.iloc[:int(datewise.shape[0]*0.95)]
valid_ml=datewise.iloc[int(datewise.shape[0]*0.95):]
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 8) 
```

```python
train_poly=poly.fit_transform(np.array(train_ml["Days Since"]).reshape(-1,1))
valid_poly=poly.fit_transform(np.array(valid_ml["Days Since"]).reshape(-1,1))
y=train_ml["total_cases"]
```

```python
linreg=LinearRegression(normalize=True)
linreg.fit(train_poly,y)
```

```python
prediction_poly=linreg.predict(valid_poly)
rmse_poly=np.sqrt(mean_squared_error(valid_ml["total_cases"],prediction_poly))
model_scores.append(rmse_poly)
print("Root Mean Squared Error for Polynomial Regression: ",rmse_poly)
```

```python
comp_data=poly.fit_transform(np.array(datewise["Days Since"]).reshape(-1,1))
plt.figure(figsize=(11,6))
predictions_poly=linreg.predict(comp_data)

fig=go.Figure()
fig.add_trace(go.Scatter(x=datewise.index, y=datewise["total_cases"],
                    mode='lines+markers',name="Train Data for total Cases"))
fig.add_trace(go.Scatter(x=datewise.index, y=predictions_poly,
                    mode='lines',name="Polynomial Regression Best Fit",
                    line=dict(color='black', dash='dot')))
fig.update_layout(title="Total Cases Polynomial Regression Prediction",
                 xaxis_title="Date",yaxis_title="Total Cases",
                 legend=dict(x=0,y=1,traceorder="normal"))
fig.show()
```

### Results
* The improved predicted values are extracted with adding polynomial features

```python
new_prediction_poly=[]
for i in range(1,18):
    new_date_poly=poly.fit_transform(np.array(datewise["Days Since"].max()+i).reshape(-1,1))
    new_prediction_poly.append(linreg.predict(new_date_poly)[0])
```
