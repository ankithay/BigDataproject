```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import csv
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv(r'C:\Users\I325229\Downloads\covid-19-world-cases-deaths-testing.csv')
df_positive_rate=df[['continent','positive_rate']]
res=df_positive_rate[df_positive_rate['positive_rate']>0]
res.positive_rate*=100
negative=100-res.positive_rate
df1=pd.DataFrame(negative)
df1.columns=['negative_rate']
final_result=res.join(df1)
final_result.style.hide_index()
f, axes = plt.subplots(figsize=(15,15))
sns.set_style('darkgrid')
fig1=sns.barplot(x="continent",y="positive_rate", data=final_result[final_result['positive_rate']>0])
f, axes = plt.subplots(figsize=(15,15))
sns.set_style('darkgrid')
fig2=sns.barplot(x="continent",y="negative_rate", data=final_result[final_result['negative_rate']>0])
```

```python
###################################################
df = pd.read_csv(r'C:\Users\I325229\Downloads\covid-19-world-cases-deaths-testing.csv')
second_question=df[['total_deaths','total_deaths_per_million','hosp_patients']]
sec1=df[['continent']]
second_res=second_question[second_question['total_deaths']>0]
re=second_res.fillna(0)
re1=re.astype(int)
re1.style.hide_index()
re1['continent']=sec1
f, axes = plt.subplots(figsize=(15,15))
sns.set_style('darkgrid')
fig3=sns.barplot(x="continent",y="total_deaths", data=re1[re1['total_deaths']>0])
f, axes = plt.subplots(figsize=(15,15))
sns.set_style('darkgrid')
fig4=sns.barplot(x="continent",y="total_deaths_per_million", data=re1[re1['total_deaths_per_million']>0])
f, axes = plt.subplots(figsize=(15,15))
sns.set_style('darkgrid')
fig5=sns.barplot(x="continent",y="hosp_patients", data=re1[re1['hosp_patients']>0])
```

```python
df = pd.read_csv(r'C:\Users\I325229\Downloads\covid-19-world-cases-deaths-testing.csv')
third_question=df[['weekly_hosp_admissions_per_million','hospital_beds_per_thousand']]
third1=df[['continent']]
third_question=third_question[third_question['weekly_hosp_admissions_per_million']>0]
re=third_question.fillna(0)
re1=re.astype(int)
re1.style.hide_index()
ne=re1.weekly_hosp_admissions_per_million/re1.hospital_beds_per_thousand
re1['continent']=third1
ne1=ne.astype(int)
df1=pd.DataFrame(ne1)
df1.columns=['beds_needed']
fin=re1.join(df1)
fin.style.hide_index()
f, axes = plt.subplots(figsize=(15,15))
sns.set_style('darkgrid')
fig6=sns.barplot(x="continent",y="beds_needed", data=fin[fin['beds_needed']>0])
```
