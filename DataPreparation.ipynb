### Imputation 
##### Imputing categorical data with most frequent value
##### Imputing numerical data with mean and ffill values

```python
mode = ''.join(df1['tests_units'].mode())
df1['tests_units'].fillna(mode, inplace=True)
df1['tests_units']
```

```python
df1.fillna(df1.mean())
```

```python
df1["total_cases"] = df1["total_cases"].fillna(df1["total_cases"].mean())
df1["total_cases"].interpolate()
```

```python
for column in df1.columns:
    percentage = df1[column].isnull().mean()
    print(f'{column}: {round(percentage*100, 2)}%')
```

```python
df1 = df1.fillna(axis = 1, method='ffill')
df1.info()
```
